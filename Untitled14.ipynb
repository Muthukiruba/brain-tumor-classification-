# 1. CHOOSE REAL IMAGE DATESET ONLINE FROM KAGGLE (BRAIN TUMOR)

```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random

def load_data(dir_list, image_size):
    # load all images in a directory
    data = []
    label = []
    image_width, image_height = image_size
    
    for directory in dir_list:
        for filename in os.listdir(directory):
            try:
                image = cv.imread(os.path.join(directory, filename), 0)
                if image is None:
                    print(f"Error: Unable to read image {os.path.join(directory, filename)}")
                    continue

                image = cv.resize(image, dsize=(image_width, image_height), interpolation=cv.INTER_CUBIC)
                image = image / 255.0
                image = image.astype(np.float32)
                data.append(image)

                if directory.endswith('yes'):
                    label.append([1])  # have tumor
                else:
                    label.append([0])  # do not have tumor
            except Exception as e:
                print(f"Error processing image {os.path.join(directory, filename)}:", e)

    data_array = np.array(data)
    label = np.array(label) 
    return data_array, label

x, y = load_data(['D:/Journal/brain tumor/archiv/yes', 'D:/Journal/brain tumor/archiv/no'], (256, 256))
# Select random indices
random_indices = random.sample(range(len(x)), 10)

# Plot randomly selected images
plt.figure(figsize=(12, 6))
for i, idx in enumerate(random_indices):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x[idx], cmap='gray')
#     plt.title('Tumor' if y[idx] == 1 else 'No Tumor')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

# 3. DESIGN AN ARCHITECTURE AND WRITE A PSEUDOCODE FOR THE PROPOSED APPROACH

```python
def split_data(test_size=0.2 , log=True):
    x_train, x_test_val, y_train, y_test_val = train_test_split(x, y, test_size=test_size,shuffle=True)
    if log==True : 
        print('Input Shape : ')
        print(x_train.shape,(y_train).shape)
        print(x_test_val.shape,(y_test_val).shape)
    x_test=x_test_val
    x_val=x_test_val
    y_test=y_test_val
    y_val=y_test_val
    
        
    x_train=x_train.reshape(-1,256,256,1)  
#     y_train=y_train.reshape(-1,100,100,1)  
#     y_test=y_test.reshape(-1,100,100,1)    
    x_test=x_test.reshape(-1,256,256,1)
    x_val=x_val.reshape(-1,256,256,1)
    if log==True : 
        print('Output Shape : ')
        print(x_train.shape,y_train.shape)
        print(x_test.shape,y_test.shape)
        print(x_val.shape,y_val.shape)
    return x_train, x_test, y_train, y_test,x_val,y_val
x_train, x_test, y_train, y_test,x_val,y_val=split_data()
validation_dataset=(x_val,y_val)
```

# 4. COMPARE THE EXISTING IMAGE FUNCTIONS AND METHODS WITH YOUR PROPOSED APPROACH.

# EXSISTING METHOD (SVM ALGORITHM)

```python

# Flatten the training and test data
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)

# Define the SVM model
svm_model = SVC(kernel='linear')

# Train the SVM model
svm_model.fit(x_train_flat, y_train)

# Predictions on the test set
y_pred_svm = svm_model.predict(x_test_flat)

# Evaluate the SVM model
accuracy = accuracy_score(y_test, y_pred_svm)
precision = precision_score(y_test, y_pred_svm)
recall = recall_score(y_test, y_pred_svm)

print("SVM Accuracy:", accuracy)
print("SVM Precision:", precision)
print("SVM Recall:", recall)
```

# PROPOSED METHOD (CNN ALGORITHM)

SVM may require relatively less training data compared to CNNs.

CNNs typically require large amounts of labeled data for training due to their high parameter count and complexity.

SVM is a simpler algorithm compared to CNNs and may struggle to model complex relationships between features in high-dimensional data like medical images.

```python
def Build_Model(input_shape=(256,256,1)):
    Model = keras.models.Sequential([
            Conv2D(16,kernel_size=(3,3),activation='relu',input_shape=input_shape),
            Conv2D(32,kernel_size=(3,3),activation='relu'),
            MaxPool2D(2,2),
            Conv2D(32,kernel_size=(3,3),activation='relu'),    
            Conv2D(32,kernel_size=(3,3),activation='relu'),    
            Conv2D(64,kernel_size=(3,3),activation='relu'),
            MaxPool2D(4,4),
            Flatten() ,    
            Dense(64,activation='relu') ,        
            Dense(32,activation='relu') ,   
            Dense(16,activation='relu'),
            Dropout(rate=0.5) ,            
            Dense(1,activation='sigmoid') ,    
            ])
    return Model
```

```python
Model=Build_Model()
Model.summary()
```

```python
Model.compile(optimizer='adam',
              loss="binary_crossentropy",
              metrics=['accuracy'])
```

```python
def make_callbacks(csv_path='my_class_log_5.csv'):
    early_stop=EarlyStopping(monitor='val_loss',patience=8,mode='min',restore_best_weights=False)
    csv_logger=CSVLogger(csv_path)
    callbacks =[early_stop,csv_logger]
    return callbacks
callbacks=make_callbacks()
```

```python
epochs = 1
Model.fit(x_train, y_train, epochs=epochs,batch_size=32,verbose=1,callbacks=callbacks,validation_data=validation_dataset)
```

# PSNR, SSIM, PRECISION, RECALL

```python
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import numpy as np

psnr_scores = []
ssim_scores = []

for i in range(len(x_test)):
    original_image = x_test[i].reshape(256, 256)
    
    # Assuming y_pred_binary contains binary predictions (0 or 1)
    binary_prediction = y_pred_binary[i]
    
    # Create an image mask based on binary prediction
    if binary_prediction == 1:
        predicted_image = np.ones_like(original_image) * 255  # Tumor present
    else:
        predicted_image = np.zeros_like(original_image)  # No tumor
    
    # Compute PSNR
    psnr_score = psnr(original_image, predicted_image)
    psnr_scores.append(-psnr_score)
    
    # Compute SSIM
    ssim_score = ssim(original_image, predicted_image)
    ssim_scores.append(ssim_score)

# Print average PSNR and SSIM scores
print("Average PSNR:", sum(psnr_scores) / len(psnr_scores))
print("Average SSIM:", sum(ssim_scores) / len(ssim_scores))
```
